```python
import numpy as np
import scipy.stats as st
from scipy.stats import expon
from scipy.stats import gamma
from scipy import stats
import scipy
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
class Film():
    def __init__(self, name, genre, imdb, director):
        print("imdb score of the film: ", imdb)
        self.name = name
        self.genre = genre
        self.imdb = imdb
        self.director = director
    def ExponentialDist(self,genre):
        population_imdb = expon.rvs(size = 1000)
        ax = sns.distplot(population_imdb, kde=True, bins=100, color='green', hist_kws={"linewidth": 15,'alpha':1})
        ax.set(xlabel = "imdb", ylabel = "Films")
        plt.show()
        np.random.seed(10) 
        mean = population_imdb.mean()
        print("Mean: ",mean," of",self.genre)
        std = population_imdb.std()
        print("Standard Deviation: ",std," of",self.genre)
        val_p = expon.pdf(population_imdb, loc=mean, scale=std)
        print("PDF: ",val_p," of ",self.genre)
        val_c = expon.cdf(population_imdb, loc=mean, scale=std)
        print("CDF: ",val_c," of",self.genre)

    def NormalDist(self,genre):
        population_imdb = norm.rvs(size = 1000, loc = 5, scale = 1)
        ax = sns.distplot(population_imdb, kde=True, bins=100, color='skyblue', hist_kws={"linewidth": 15,'alpha':1})
        ax.set(xlabel = "imdb", ylabel = "Films")
        plt.show()
        mean = population_imdb.mean()
        print("Mean: ",mean," of",self.genre)
        std = population_imdb.std()
        print("Standard Deviation: ",std," of",self.genre)
        val_p = stats.norm(mean,std).pdf(0)
        print("PDF: ",val_p," of ",self.genre)
        val_c = stats.norm.cdf(population_imdb)
        print("CDF: ",val_c," of",self.genre)
```

```python
film_1 = Film("Avengers:Endgame","Sci-fi",8.6,"Anthony Russo")
film_2 = Film("Once Upon a Time in Hollywood","Comedy",8.1,"Quentin Tarantino")
film_3 = Film("The Green Mile","Drama",8.6,"Frank Darabont")
film_4 = Film("Lord of The Rings:The Fellowship of the Ring","Fantastic",8.8,"Peter Jackson")
```

```python
film_1.ExponentialDist(film_1.genre) #imdb score for the films that have the same genres with film_1
```
