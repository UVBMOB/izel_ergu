```python
def describeColumn(columnName): # her column için hesaplanması gereken değerler
    print("Mean:", bank[columnName].mean())
    print("Median: ", bank[columnName].median())
    print("Mode: ", bank[columnName].mode())
    print("Standard Deviation: ",bank[columnName].std())
    print("Sum: ", bank[columnName].sum())
    print("Count: ", bank[columnName].count())
    print("Min: ", bank[columnName].min())
    print("Max: ", bank[columnName].max())
    print("Abs: ", bank[columnName].abs())
    print("Prod: ", bank[columnName].prod())
    print("Cum sum : ", bank[columnName].cumsum())
    print("Cum prod: ", bank[columnName].cumprod())
# sadece continuous değerler için describeColumn fonksiyonu kullanıldı
```

```python
# SHAPIRO
# varsayım kontrolü
from scipy.stats import shapiro
def shapiroTest(columnName):
    stat, p  = shapiro(bank[columnName]) # dataframe yapmadan ve isim vermeden önceki hali
    print("Statistics=%.3f, p=%.3f" % (stat,p))
    alpha = 0.05
    if p > alpha:
        print("Örneklem gaussian (normal) dağılımından gelmektedir (fail to reject)")
    else:
        print("Örneklem gaussian (normal) dağılımından gelmemektedir (reject H0)")
        
for column in bank.columns:
    if(column != 'reason' and column != 'job' and column != 'bad'):
        print("--------------------")
        print("for column ",column)
        shapiroTest(column)
        print("--------------------")

# İlk olarak columnların normal dağılımdan gelip gfelmediğini kontrol ediyoruz
```

```python
from scipy.stats import levene
# LEVENE
def leveneTest(column):
    stat, p = levene(bank[column][bank['bad'] == 0], bank[column][bank['bad'] == 1]) #equal variance test
    print("Statistics=%.3f, p=%.3f" % (stat,p))
    alpha = 0.05
    if p > alpha:
        print("Örneklemler sabit varyansa sahiptir (fail to reject)")
    else:
        print("Örneklemler sabit varyansa sahip değildir (reject H0)")
        
for column in bank.columns:
    if(column != 'reason' and column != 'job' and column != 'bad'):
        print("--------------------")
        print("for column ",column)
        leveneTest(column)
        print("--------------------")
```

```python
# Hangi değişkenler target değeri etkiliyor.
# h0: beta0 = beta1 = ... = 0
# h1: En az biri farklı
from scipy.stats import f_oneway 
# ANOVA
def onewayTest(column):
    stat, p = f_oneway(bank[column][bank['bad'] == 0], bank[column][bank['bad'] == 1]) #equal variance test
    print("Statistics=%.3f, p=%.3f" % (stat,p))
    alpha = 0.05
    if p > alpha:
        print("Örneklemler sabit varyansa sahiptir (fail to reject)")
    else:
        print("Örneklemler sabit varyansa sahip değildir (reject H0)")
        
for column in bank.columns:
    if(column != 'reason' and column != 'job' and column != 'bad'):
        print("--------------------")
        print("for column ",column)
        onewayTest(column)
        print("--------------------")

# value, yoj, derog, delinq, clage, ninq and deptinc columnları kötü müşteriyi tanımlamak için ayrıştırıcı --> reject h0
```
